```python
import seaborn as sns
```

```python
df = sns.load_dataset('titanic')
df.head()
```

```python
df['decade'] = [age//10*10 for age in df['age']]
df.head()
```

### Visualization

> Relplot - access to both scatter and line plot

```python
#sns.set_style('white')
```

```python
sns.relplot(x='age', y='fare', data=df);
```

>Categorical variable: hue, style and size

```python
sns.relplot(x='age', y='fare', data=df, hue='pclass');
```

```python
sns.relplot(x='age', y='fare', data=df, hue='pclass', style='pclass');
```

```python
sns.relplot(x='age', y='fare', data=df, hue='pclass', style='pclass', size='pclass');
```

```python
sns.relplot(x='age', y='fare', data=df, kind='line'); #the mean value of the aggregated passengers
```

```python
sns.relplot(x='decade', y='fare', data=df, kind='line'); # shows the age group between 50 and 60 paid more fare
```

```python
sns.relplot(x='decade', y='fare', data=df, kind='line', hue='pclass');
```

```python
sns.relplot(x='decade', y='fare', data=df, kind='line', hue='pclass', style='pclass');
```

> Small mutiples (Facet Grid)

```python
sns.relplot(x='age', y='fare', data=df, col='pclass');
```

```python
sns.relplot(x='age', y='fare', data=df, row='pclass');
```

```python
sns.relplot(x='decade', y='fare', data=df, kind='line', col='pclass');
```

```python
sns.relplot(x='age', y='fare', data=df, col='deck');
```

```python
sns.relplot(x='age', y='fare', data=df, col='deck', col_wrap=3);
```

### Styling

> Additional keyword arguments will be passed to the kind of plot selected.

```python
sns.relplot(x='age', y='fare', data=df, col='pclass', hue='survived');
```

```python
sns.relplot(x='age', y='fare', data=df, col='pclass', hue='survived', s=150);
```

```python
sns.relplot(x='age', y='fare', data=df, col='pclass', hue='survived',
            s=150, palette=['black','red']);
```

```python
sns.relplot(x='age', y='fare', data=df, col='pclass', hue='survived',
            s=150, palette=['black','red'], marker='^');
```

```python
sns.relplot(x='age', y='fare', data=df, col='pclass', hue='survived',
            s=150, palette=['black','red'], style='survived');
```

```python
sns.relplot(x='age', y='fare', data=df, col='pclass', hue='survived',
            s=150, palette=['black','red'], style='survived', markers=['X','o']);
```

```python
sns.relplot(x='decade', y='survived', data=df, kind='line');
```

```python
sns.relplot(x='decade', y='survived', data=df[df['age']<70], kind='line');
```

```python
sns.relplot(x='decade', y='survived', data=df[df['age']<70], kind='line', hue='sex');
```

```python
sns.relplot(x='decade', y='survived', data=df[df['age']<70], kind='line', hue='sex', ci=None);
```

```python
g=sns.relplot(x='decade', y='survived', data=df[df['age']<70], kind='line', hue='sex', ci=None);
type(g)
```

```python
g=sns.relplot(x='decade', y='survived', 
              data=df[df['age']<70], 
              kind='line', 
              hue='sex',
              ci=None);
g.set_xlabels('Passenger Age Group');
```

```python
g=sns.relplot(x='decade', y='survived', 
              data=df[df['age']<70], 
              kind='line', 
              hue='sex',
              ci=None);
g.set_xlabels('Passenger Age Group');
g.set_xlabels('Percent Survived');
```

```python
g=sns.relplot(x='decade', y='survived', 
              data=df[df['age']<70], 
              kind='line', 
              hue='sex',
              ci=None,
              col='pclass');
g.set_xlabels('Passenger Age Group');
g.set_xlabels('Percent Survived');
```
