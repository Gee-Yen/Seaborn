```python
import seaborn as sns
```

```python
sns.__version__
```

```python
penguins = sns.load_dataset('penguins')
penguins.head()
```

```python
penguins.shape
```

```python
penguins.isna().sum()
```

```python
penguins.dropna(inplace=True)
```

```python
penguins.shape
```

## Intro Visuals

```python
sns.set_style('darkgrid')
```

> <b> Numerical variables </b>

```python
sns.histplot(x='bill_length_mm', data=penguins);
```

```python
sns.histplot(y='bill_length_mm', data=penguins);
```

```python
sns.histplot(x='bill_length_mm', data=penguins,kde=True);
```

```python
sns.histplot(x='bill_length_mm', data=penguins, bins=20);
```

```python
sns.histplot(x='bill_length_mm', data=penguins, bins=[30,40,45,50,60]);
```

### binwidth and binrange

```python
sns.histplot(x='bill_length_mm', data=penguins, binwidth=10)
```

```python
sns.histplot(x='bill_length_mm', data=penguins, binwidth=10, binrange=(30,60))
```

## Statistics

### stat

```python
sns.histplot(x='bill_length_mm', data=penguins, stat='count');
```

> <b>density</b> - Area of histogram sums to one

```python
sns.histplot(x='bill_length_mm', data=penguins, stat='density');
```

> <b>probability</b> - Height of histogram bars to one

```python
sns.histplot(x='bill_length_mm', data=penguins, stat='probability');
```

### cumualtive

```python
sns.histplot(x='bill_length_mm', data=penguins, stat='probability', 
             element='step', fill=False);
```

```python
sns.histplot(x='bill_length_mm', data=penguins, stat='probability', 
             element='step', fill=False, cumulative=True);
```

> <b> Categorical variables </b>

```python
sns.histplot(x='bill_length_mm', data=penguins, hue='species');
```

```python
sns.histplot(x='bill_length_mm', data=penguins, hue='species', element='step');
```

```python
sns.histplot(x='bill_length_mm', data=penguins, hue='species', element='poly');
```

```python
sns.histplot(x='bill_length_mm', data=penguins, hue='species', multiple='stack');
```

```python
sns.histplot(x='bill_length_mm', data=penguins, hue='species', multiple='fill');
```

### Bivariate histplot

```python
sns.set_style('white')
```

```python
sns.histplot(x='bill_length_mm', y='bill_depth_mm', data=penguins);
```

```python
sns.histplot(x='bill_length_mm', y='bill_depth_mm', data=penguins, cbar=True);
```

```python
sns.histplot(x='bill_length_mm', y='bill_depth_mm', data=penguins, hue='species');
```

#### Discrete color map

```python
sns.histplot(x='bill_length_mm', y='species', data=penguins, hue='species');
```

### Styling

```python
sns.histplot(x='species', data=penguins, hue='sex');
```

```python
sns.histplot(x='species', data=penguins, hue='sex', multiple='dodge');
```

```python
sns.histplot(x='species', data=penguins, hue='sex', multiple='dodge', shrink=0.8);
```

```python
sns.histplot(x='species', data=penguins, color='indigo');
```

```python
sns.histplot(x='species', data=penguins, hue='sex', multiple='dodge', shrink=0.8, palette='bone');
```

```python
sns.histplot(x='bill_length_mm', data=penguins, fill=False);
```
