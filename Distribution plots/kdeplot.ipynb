### KDE plot

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars = sns.load_dataset('mpg').dropna()
cars.shape
```

```python
cars.head()
```

#### Univariate 

```python
sns.kdeplot(cars.horsepower)
```

```python
sns.kdeplot(cars.horsepower, shade=True);
```

```python
sns.kdeplot(cars.horsepower, bw_adjust=1);
```

```python
sns.kdeplot(cars.horsepower, cumulative=True);
```

#### Bivariate

```python
sns.kdeplot(x=cars.horsepower, y=cars.mpg);
```

```python
sns.kdeplot(x=cars.horsepower, y=cars.mpg, n_levels=20);
```

```python
sns.kdeplot(x=cars.horsepower, y=cars.mpg, shade=True);
```

```python
sns.kdeplot(x=cars.horsepower, y=cars.mpg, shade=True, thresh=0.05);
```

```python
sns.kdeplot(x=cars.horsepower, y=cars.mpg, shade=True, thresh=0.05, cbar=True);
```

```python
sns.kdeplot(x=cars.horsepower, y=cars.mpg, hue=cars.cylinders);
```
